UPLOADING DATASET

```python
import pandas as pd
data = pd.read_csv('/content/Advertising.csv')
```

DATA PREPROCESSING

```python
data
```

```python
import matplotlib.pyplot as plt
plt.scatter(data['Radio'], data['Sales'])
plt.xlabel('Radio')
plt.ylabel('Sales')
plt.show()
```

```python
data.isnull().sum()
```

TRAIN-TEST SPLITTING OF DATA AND SCALING

```python
x = data.iloc[: ,[1,2,3]]
y = data.iloc[:, [4]]
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.3, random_state=50)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
xtrain_scaled = scaler.fit_transform(xtrain)
xtest_scaled = scaler.transform(xtest)
```

Flattening ytrain and ytest array to 1D

```python
import numpy as np
```

```python
ytrain = np.asarray(ytrain)
ytrain_flat  = ytrain.flatten()
ytrain_flat.ndim
```

```python
ytest = np.asarray(ytest)
ytest_flat = ytest.flatten()
ytest_flat.ndim
```

APPLYING RANDOM FOREST REGRESSOR

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor()
regressor.fit(xtrain_scaled, ytrain_flat)
ypred = regressor.predict(xtest_scaled)
```

```python
ypred
```

EVALUATION OF MODEL

```python
import numpy as np
from sklearn import metrics
score = metrics.r2_score(ytest_flat,ypred)
print("The R2 score is ", score)
```

```python
import matplotlib.pyplot as plt
plt.title("ytest vs ypred", fontsize=20)
plt.plot(ypred, color="blue")
plt.plot(ytest, color="red")
plt.legend(labels = ('ypred', 'ytest'))
```

TESTING RANDOM SAMPLE FROM TEST DATA

```python
inp = [[228.0, 37.7, 32.0]]
scaled_inp = scaler.transform(inp)
result = regressor.predict(scaled_inp)
print("Your sales value is ",result)
print("Actual sales value is ", ytest[8])
```

You can check random values by passing into below function

```python
def sales(tv, radio, newspaper):
  inp = [[tv, radio, newspaper]]
  scaled_inp = scaler.transform(inp)
  result = regressor.predict(scaled_inp)
  print(f"Sales value is {result}")
```
